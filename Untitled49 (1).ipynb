```python
import os, math, time
import numpy as np, pandas as pd
import torch, torch.nn as nn, torch.nn.functional as F
import matplotlib.pyplot as plt
import yfinance as yf
import pandas_datareader.data as pdr
from torch_geometric.nn import GCNConv

DEVICE = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print('Using:', DEVICE)
```

```python
def load_real_data(start='2006-01-01', end=None):
    if end is None:
        end = pd.Timestamp.today().strftime('%Y-%m-%d')

    fred_series = {
        'VIXCLS':'VIX',
        'FEDFUNDS':'FedFunds',
        'UNRATE':'Unemployment',
        'CPIAUCSL':'CPI',
        'M2SL':'M2'
    }

    sp = yf.download('^GSPC', start=start, end=end, progress=False)['Close']
    sp.name = 'SP500'

    macro_df = pd.DataFrame()
    for k,n in fred_series.items():
        try:
            s = pdr.DataReader(k,'fred',start,end)
            macro_df[n] = s.iloc[:,0]
        except Exception as e:
            print('FRED fail',k, e)

    data = pd.concat([sp, macro_df], axis=1).dropna()
    print('Data shape:', data.shape)
    return data

data = load_real_data()
```

```python
def make_node_features_from_timeseries(data, feature_window=16):
    df = data.copy()
    df = df.pct_change().fillna(0.0)

    df_std = (df - df.mean())/(df.std()+1e-12)

    last = df_std.iloc[-feature_window:]
    X = np.vstack([last[c].values.astype(np.float32) for c in df_std.columns])

    return torch.tensor(X, dtype=torch.float32), df_std

X_init, df_std = make_node_features_from_timeseries(data, 16)
print('X shape:', X_init.shape)
```

```python

def build_corr_graph(df_std, threshold=0.35):
    corr = np.corrcoef(df_std.values.T)
    N = corr.shape[0]
    edges=[]

    for i in range(N):
        for j in range(N):
            if i!=j and abs(corr[i,j])>=threshold:
                edges.append([i,j])

    if len(edges)==0:
        edges=[[i,j] for i in range(N) for j in range(N) if i!=j]

    return torch.tensor(edges,dtype=torch.long).T.contiguous().to(DEVICE)

edge_index = build_corr_graph(df_std)
print('Edges:', edge_index.shape)
```

```python
EMBED_DIM = X_init.shape[1]
NUM_CLASSES = 2

class GNNTeacher(nn.Module):
    def __init__(self):
        super().__init__()
        self.conv1 = GCNConv(EMBED_DIM, 256)
        self.conv2 = GCNConv(256, NUM_CLASSES)

    def forward(self, x, edge_index):
        h = F.relu(self.conv1(x, edge_index))
        return self.conv2(h, edge_index)

class MLPStudent(nn.Module):
    def __init__(self):
        super().__init__()
        self.fc1 = nn.Linear(EMBED_DIM, 256)
        self.fc2 = nn.Linear(256, NUM_CLASSES)

    def forward(self, x):
        return self.fc2(F.relu(self.fc1(x)))

teacher = GNNTeacher().to(DEVICE)
student = MLPStudent().to(DEVICE)
X_init = X_init.to(DEVICE)
```

```python
def simulate_price(logits):
    p = F.softmax(logits, dim=1)
    return (p[:,1]-p[:,0]).cpu().numpy()

def run_twin_market_trial(T=240, crisis_start=60, crisis_end=180):

    X_A = X_init.clone()  # baseline market
    X_B = X_init.clone()  # shocked market

    gA, gB = [], []
    mA, mB = [], []

    for t in range(T):

        # Normal noise
        sigma = 0.005

        # Crisis regime shift
        if crisis_start <= t < crisis_end:
            sigma = 0.03

        # Baseline market A
        X_A = X_A + sigma * torch.randn_like(X_A)

        # Shocked market B
        X_B = X_B + sigma * torch.randn_like(X_B)

        if t == crisis_start:
            k = max(1, int(0.02 * X_B.shape[0]))
            idx = torch.randperm(X_B.shape[0])[:k].to(DEVICE)
            X_B[idx] += 2.0 * torch.randn_like(X_B[idx])  # sentiment shock

        if crisis_start <= t < crisis_end:
            X_B = X_B * (1 + 0.02 * torch.randn_like(X_B))  # liquidity shock

        with torch.no_grad():
            gA.append(simulate_price(teacher(X_A, edge_index)).mean())
            gB.append(simulate_price(teacher(X_B, edge_index)).mean())

            mA.append(simulate_price(student(X_A)).mean())
            mB.append(simulate_price(student(X_B)).mean())

    return np.array(gA), np.array(gB), np.array(mA), np.array(mB), crisis_start, crisis_end
```

```python
gA, gB, mA, mB, cs, ce = run_twin_market_trial()

plt.figure(figsize=(10,5))
plt.plot(gA, label='GNN Market A')
plt.plot(gB, label='GNN Market B (Shocked)')
plt.plot(mA, linestyle='--', label='MLP Market A')
plt.plot(mB, linestyle='--', label='MLP Market B (Shocked)')
plt.axvspan(cs, ce, alpha=0.12, color='red', label='Crisis')
plt.legend()
plt.grid(alpha=0.25)
plt.title('EXP-2: Twin Market Shock Stress Test')
plt.show()
```
