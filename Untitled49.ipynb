```python
import os, math, time
import numpy as np, pandas as pd
import torch, torch.nn as nn, torch.nn.functional as F
import matplotlib.pyplot as plt
import yfinance as yf
import pandas_datareader.data as pdr
DEVICE = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print('Using:', DEVICE)

```

```python
def load_real_data(start='2006-01-01', end=None):
    if end is None: end = pd.Timestamp.today().strftime('%Y-%m-%d')
    fred_series = {'VIXCLS':'VIX','FEDFUNDS':'FedFunds','UNRATE':'Unemployment','CPIAUCSL':'CPI','M2SL':'M2'}
    sp = yf.download('^GSPC', start=start, end=end, progress=False)['Close']
    sp.name = 'SP500' # Directly assign the name to the Series
    macro_df = pd.DataFrame()
    for k,n in fred_series.items():
        try:
            s = pdr.DataReader(k,'fred',start,end)
            macro_df[n] = s.iloc[:,0]
        except Exception as e:
            print('FRED fail',k, e)
    data = pd.concat([sp, macro_df], axis=1).dropna()
    print('Data shape:', data.shape)
    return data

data = load_real_data()
```

```python
def make_node_features_from_timeseries(data, feature_window=16):
    df = data.copy()
    df['^GSPC'] = df['^GSPC'].pct_change().fillna(0.0)
    df_std = (df - df.mean())/(df.std()+1e-12)
    last = df_std.iloc[-feature_window:]
    X = np.vstack([last[c].values.astype(np.float32) for c in df_std.columns])
    return torch.tensor(X, dtype=torch.float32), df_std

X_init, df_std = make_node_features_from_timeseries(data, 16)
print('X shape:', X_init.shape)
```

```python
def build_corr_graph(df_std, threshold=0.35):
    corr = np.corrcoef(df_std.values.T)
    N = corr.shape[0]
    edges=[]
    for i in range(N):
        for j in range(N):
            if i!=j and abs(corr[i,j])>=threshold:
                edges.append([i,j])
    if len(edges)==0:
        edges=[[i,j] for i in range(N) for j in range(N) if i!=j]
    return torch.tensor(edges,dtype=torch.long).T.contiguous().to(DEVICE)

edge_index = build_corr_graph(df_std)
print('Edges:', edge_index.shape)
```

```python
import torch.nn as nn
import torch.nn.functional as F
from torch_geometric.nn import GCNConv

# GNN Teacher model definition from exp_full.py
class GNNTeacher(nn.Module):
    def __init__(self):
        super().__init__()
        self.conv1 = GCNConv(EMBED_DIM, 256)
        self.conv2 = GCNConv(256, NUM_CLASSES)

    def forward(self, x, edge_index):
        h = F.relu(self.conv1(x, edge_index))
        return self.conv2(h, edge_index)

# MLP Student model definition from exp_full.py
class MLPStudent(nn.Module):
    def __init__(self):
        super().__init__()
        self.fc1 = nn.Linear(EMBED_DIM, 256)
        self.fc2 = nn.Linear(256, NUM_CLASSES)

    def forward(self, x):
        return self.fc2(F.relu(self.fc1(x)))
```

```python
import torch.nn as nn
import torch.nn.functional as F
from torch_geometric.nn import GCNConv

EMBED_DIM = X_init.shape[1]   # <-- auto-match features
NUM_CLASSES = 2

class GNNTeacher(nn.Module):
    def __init__(self):
        super().__init__()
        self.conv1 = GCNConv(EMBED_DIM, 256)
        self.conv2 = GCNConv(256, NUM_CLASSES)

    def forward(self, x, edge_index):
        h = F.relu(self.conv1(x, edge_index))
        return self.conv2(h, edge_index)

class MLPStudent(nn.Module):
    def __init__(self):
        super().__init__()
        self.fc1 = nn.Linear(EMBED_DIM, 256)
        self.fc2 = nn.Linear(256, NUM_CLASSES)

    def forward(self, x):
        return self.fc2(F.relu(self.fc1(x)))
```

```python
# Instantiate the GNNTeacher and MLPStudent models
teacher = GNNTeacher().to(DEVICE)
student = MLPStudent().to(DEVICE)
X_init = X_init.to(DEVICE)
```

```python
def simulate_price(logits):
    p=F.softmax(logits,dim=1); return (p[:,1]-p[:,0]).cpu().numpy()
def max_drawdown_level(s):
    peak=s[0]; m=0
    for x in s:
        peak=max(peak,x); m=max(m, peak-x)
    return m
def recovery_time_from_shock(d,eps=0.05):
    mx=np.max(np.abs(d)); th=eps*mx
    for i,x in enumerate(d):
        if abs(x)<=th: return i
    return len(d)

def run_dynamic_trial(T=240, crisis_start=60, crisis_end=180):
    X_t=X_init.clone(); g=[]; m=[]
    for t in range(T):
        sigma=0.005 if not (crisis_start<=t<crisis_end) else 0.03
        X_t = X_t + sigma*torch.randn_like(X_t)
        if t==crisis_start:
            k=max(1,int(0.02*X_t.shape[0])); idx=torch.randperm(X_t.shape[0])[:k].to(DEVICE)
            X_t[idx]+=2.0*torch.randn_like(X_t[idx])
        if crisis_start<=t<crisis_end:
            X_t = X_t*(1+0.01*torch.randn_like(X_t))
        with torch.no_grad():
            g.append(simulate_price(teacher(X_t,edge_index)).mean())
            m.append(simulate_price(student(X_t)).mean())
    return np.array(g), np.array(m), crisis_start, crisis_end
```

```python
g,m,cs,ce = run_dynamic_trial()
plt.figure(figsize=(10,4))
plt.plot(g,label='GNN'); plt.plot(m,label='MLP')
plt.axvspan(cs,ce,alpha=0.12,color='red',label='crisis')
plt.legend(); plt.grid(alpha=0.25)
plt.title('Exp-2 Dynamic Stress Test')
plt.show()
```
